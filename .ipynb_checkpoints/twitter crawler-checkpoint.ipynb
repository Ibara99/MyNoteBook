```python
import sqlite3
from sqlite3 import Error
 
           
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return conn
def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement
    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)

def duplicate_tweet(conn, tweet):
    """
    Query tasks by priority
    :param conn: the Connection object
    :param priority:
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM tweets WHERE user=? AND tgl=? AND tweet=? AND favorite=?", tweet)
 
    rows = cur.fetchall()
 
    if len(rows) > 0:
        return True
    else: 
        return False
    
def insert_tweet(conn, tweet):
    """
    Create a new project into the projects table
    :param conn:
    :param tweet:
    :return: project id
    """
    if not duplicate_tweet(conn, tweet):
        sql = ''' INSERT INTO tweets(user, tgl, tweet, favorite)
                  VALUES(?,?,?,?) '''
        cur = conn.cursor()
        cur.execute(sql, tweet)
        conn.commit();
        return cur.lastrowid
    else:
        print("tweet duplicated")
        cur = conn.cursor()
        return cur.lastrowid

def select_all_tweets(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM tweets")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


db_name = 'tes3.db'
create_connection(db_name)
sql_tabel_tweet = """ CREATE TABLE IF NOT EXISTS tweets (
                                        id integer PRIMARY KEY AUTOINCREMENT,
                                        user text NOT NULL,
                                        tgl text NOT NULL,
                                        tweet text NOT NULL,
                                        favorite integer NOT NULL
                                    ); """
# create a database connection
conn = create_connection(db_name)

# create tables
if conn is not None:
    # create table
    create_table(conn, sql_tabel_tweet)
    tes_tweet = ("ibara","18 Maret 2020", "hellow world!", 3)
    insert_tweet(conn,tes_tweet)
    #select_all_tweets(conn)
else:
    print("Error! cannot create the database connection.")
#conn.close()
```

```python
from twython import TwythonStreamer
import csv, sys

# Filter out unwanted data
def process_tweet(tweet):
    d = {}
    d['hashtags'] = [hashtag['text'] for hashtag in tweet['entities']['hashtags']]
    try:
        #jika tweet aslinya panjang
        d['text'] = tweet['extended_tweet']['full_text']
        d['user'] = tweet['user']['screen_name']
        d['date'] = tweet['created_at']
        d['favorite'] = tweet['favorite_count']
        print('tess')
    except:
        print('tes')
        try: #jika ada retweet yg panjang
            d['text'] = tweet['retweeted_status']['extended_tweet']['full_text']
            d['user'] = tweet['retweeted_status']['user']['screen_name']
            d['date'] = tweet['retweeted_status']['created_at']
            d['favorite'] = tweet['retweeted_status']['favorite_count']
            print("tesss")
        except:
            d['text'] = tweet['text']
            d['user'] = tweet['user']['screen_name']
            d['date'] = tweet['created_at']
            d['favorite'] = tweet['favorite_count']
    return d
def tweet_format(t):
    return (t['user'], t['date'], t['text'], t['favorite'])
    
    
# Create a class that inherits TwythonStreamer
class MyStreamer(TwythonStreamer):     
    data=None
    # Received data
    def on_success(self, data):

        # Only collect tweets in English
        try:
            self.data=data
            if data['lang'] == 'in':
                tweet_data = process_tweet(data)
                #print(data)
                self.save_to_csv(tweet_data)
        except:
            print(sys.exc_info()[0])

    # Problem with the API
    def on_error(self, status_code, data):
        print(status_code, data)
        self.disconnect()
        
    # Save each tweet to csv file
    def save_to_csv(self, tweet):
        t = tweet_format(tweet)
        print(t)
        conn = create_connection(db_name)
        insert_tweet(conn, t)
        conn.close()
        '''
        with open(r'saved_tweets.csv', 'a', encoding='utf-8') as file:
        
            writer = csv.writer(file)
            writer.writerow(list(tweet.values()))
        '''
        
            
```

```python
import pandas as pd
import numpy as np
import sqlite3


def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)


def get_all_tweets(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM tweets")
 
    return cur.fetchall()

db_name = 'tes3.db'
# create a database connection
conn = create_connection(db_name)
data = np.array(get_all_tweets(conn))
df = pd.DataFrame(data[:,1:], data[:, 0], ["user", "date", "tweet", "favorite_count"])
```
